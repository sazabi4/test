```python
import corner
import numpy as np
from start_plot import *
import matplotlib.pyplot as plt
from astropy import units as u
from astropy import constants as const
import os
%matplotlib inline
```

```python
import bovy_coords

from astropy.cosmology import WMAP7 

def history_search(haloname, ID, tag):
    temp = np.genfromtxt('/Users/astrolab/Dropbox/dwarfgalaxy/docs/IsolatedTrees/'+haloname+'/'+tag+'.txt')
    idlist = np.genfromtxt('/Users/astrolab/Dropbox/dwarfgalaxy/docs/IsolatedTrees/'+haloname+'/'+'ID.txt', dtype=int)
    idx = idlist[:,0] == ID
    return temp[idx][0][:-5]

def history_search_host(hostname, tag):
    temp = np.genfromtxt('/Users/astrolab/Dropbox/dwarfgalaxy/docs/IsolatedTrees/'+hostname+'/'+tag+'.txt')
    return temp[0][:-5]
    
def find_infall_time(t, rvir, d):
    delta = d - rvir
    for i in range(len(t)):
        if delta[::-1][i] < 0:
            break
    return 0.5*(t[::-1][i-1]+t[::-1][i])

def find_orbit_parameter(t, rvir, d):
    delta = d - rvir
    for i in range(len(t)):
        if delta[::-1][i] < 0:
            break
    it = t[::-1][i]
    rp = d[0:len(d)-i].min()
    ap = d[0:len(d)-i].max()
    rt = t[d == rp]
    at = t[d == ap]
    return rp, ap, rt[0], at[0]


def plot_orbit(haloname, ID):

    a = history_search(haloname, ID, 'scale')
    x = history_search(haloname, ID, 'X') *1000*u.kpc #*a
    y = history_search(haloname, ID, 'Y') *1000*u.kpc #*a
    z = history_search(haloname, ID, 'Z') *1000*u.kpc #*a
    a0 = history_search_host(haloname, 'scale')
    x0 = history_search_host(haloname, 'X') *1000*u.kpc #*a
    y0 = history_search_host(haloname, 'Y') *1000*u.kpc #*a
    z0 = history_search_host(haloname, 'Z') *1000*u.kpc #*a
    rvir = history_search_host(haloname, 'Rvir') *u.kpc #*a
    redshift = (1/a - 1)
    t = WMAP7.lookback_time(redshift)
    d = np.sqrt((x-x0)**2+ (y-y0)**2+ (z-z0)**2)

    infall_time = find_infall_time(t, rvir, d)
    rp, ap, rt, at = find_orbit_parameter(t, rvir, d)
  
    print 'Host Halo: %s'%haloname
    print 'Subhalo ID: %i'%ID
    print 'Infall time = %5.2f Gyr'%np.array(infall_time)
    print 'Period = %5.2f Gyr'%np.array(abs(at-rt)*2)
    print 'Apocenter: %5.2f kpc'%np.array(ap)
    print 'Pericenter: %5.2f kpc'%np.array(rp)
    print 'Eccentricity: %5.2f'%np.array((ap-rp)/(ap+rp))

    mask = (np.array(t) < 5)
    plt.figure(figsize = [10,10])
    plt.subplot(221)
    plt.plot(t[mask], d[mask],'o', label='distance to host')
    plt.plot(t[mask], rvir[mask],'rs', label='Host Rvir')
    plt.axvline(np.array(infall_time), color='k', linestyle='--')
    plt.axvline(np.array(rt), color='y', linestyle='--')
    plt.axvline(np.array(at), color='y', linestyle='--')
    plt.legend(loc=1)
    plt.ylim(0,400)
    plt.xlim(-1, 5)
    plt.xlabel('Lookback time (Gyr)')
    plt.ylabel('Comoving Distance (Kpc)')

    plt.subplot(222)
    plt.plot(x[mask]-x0[mask], y[mask]-y0[mask], 'o')
    plt.plot(x[mask]-x0[mask], y[mask]-y0[mask], '-r')
    plt.plot(0,0,'y*',ms=30)
    plt.xlim(-200,200)
    plt.ylim(-200,200)
    plt.xlabel('Comoving Distance in X (Kpc)')
    plt.ylabel('Comoving Distance in Y (Kpc)')

    plt.subplot(223)
    plt.plot(x[mask]-x0[mask], z[mask]-z0[mask], 'o')
    plt.plot(x[mask]-x0[mask], z[mask]-z0[mask], '-r')
    plt.plot(0,0,'y*',ms=30)
    plt.xlim(-200,200)
    plt.ylim(-200,200)
    plt.xlabel('Comoving Distance in X (Kpc)')
    plt.ylabel('Comoving Distance in Z (Kpc)')
    
    plt.subplot(224)
    plt.plot(y[mask]-y0[mask], z[mask]-z0[mask], 'o')
    plt.plot(y[mask]-y0[mask], z[mask]-z0[mask], '-r')
    plt.plot(0,0,'y*',ms=30)
    plt.xlim(-200,200)
    plt.ylim(-200,200)
    plt.xlabel('Comoving Distance in Y (Kpc)')
    plt.ylabel('Comoving Distance in Z (Kpc)')

    plt.tight_layout()
    #plt.savefig('elvis/'+str(ID)+'.eps')
    plt.show()
    return np.array(infall_time), np.array(abs(at-rt)*2), np.array(ap), np.array(rp), np.array((ap-rp)/(ap+rp))

def find_halo(d, vr, dd, dvr, m):
    count = 0
    f = open('elvis/Eri2_analog_subhalos.txt','w')
    f.writelines('#Index HostHalo HostHaloID  SubHaloID SubHaloMvir(z=0) SubHaloStellarMass(z=0)  vtotal(km/s)  \
    vlos(km/s)  distance(kpc)  SubHaloPID  InfallTime(Gyr)  Period(Gyr) ApoCenter(Gyr)  PeriCenter(Gyr)  Eccentricity(ap-rp/ap+rp)\n ')
    for i in os.listdir('/Users/astrolab/Dropbox/dwarfgalaxy/docs/IsolatedCatalogs/')[0:]:
        if i[0] == 'i':
            #print i
            halo = np.genfromtxt('/Users/astrolab/Dropbox/dwarfgalaxy/docs/IsolatedCatalogs/'+i,  dtype=None, names=True)
            x0 = halo['XMpc'][0] * 1000
            y0 = halo['YMpc'][0] * 1000
            z0 = halo['ZMpc'][0] * 1000
            vx0 = halo['Vxkms'][0]
            vy0 = halo['Vykms'][0]
            vz0 = halo['Vzkms'][0]
            hostID = halo['ID'][0]
            l,b,dsim = bovy_coords.XYZ_to_lbd(halo['XMpc']*1000-x0, halo['YMpc']*1000-y0, halo['ZMpc']*1000-z0, degree=True).T
            vsim,pmll,pmbb = bovy_coords.vxvyvz_to_vrpmllpmbb(halo['Vxkms'] - vx0, halo['Vykms'] - vy0, halo['Vzkms'] -vz0, l, b, dsim, degree=True).T
            #plt.hist(vsim)
            v3d = np.sqrt((halo['Vxkms'] - vx0)**2+(halo['Vykms'] - vy0)**2+(halo['Vzkms'] -vz0)**2)
            mass = halo['Mstar_preferredMsun']
            mask = (vsim > vr - 3*dvr) & (vsim < vr + 3*dvr) & (dsim > d - 3*dd ) & (dsim < d + 3*dd) & (mass > m/1000) & (mass < m * 1000)
            analog = halo[mask]
            hostname = i.split('.')[0]
            for i in range(len(analog)):
                print count, hostname, hostID, analog[i]['ID'], analog[i]['MvirMsun'],analog[i]['Mstar_preferredMsun'], v3d[mask][i],vsim[mask][i],dsim[mask][i], analog[i]['PID']
                infall_time, period, ap, rp, ecc = plot_orbit(hostname, analog[i]['ID'])
                f.writelines('%2i %10s %10i %10i %15.5g %15.5g %6.1f %6.1f %6.1f %10i %5.2f %5.2f %7.2f %7.2f %4.2f\n'\
                           %(count, hostname, hostID, analog[i]['ID'], analog[i]['MvirMsun'], analog[i]['Mstar_preferredMsun'], \
                             v3d[mask][i], vsim[mask][i], dsim[mask][i], analog[i]['PID'], \
                             infall_time, period, ap, rp, ecc))
                
                count = count+1
    f.close()

            
find_halo(23, -195, 5, 10, 3000)
```
