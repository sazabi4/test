```python
import pyfits
import matplotlib.pyplot as plt
import numpy as np
from start_plot import *
from matplotlib.ticker import ScalarFormatter, FormatStrFormatter
%matplotlib inline
```

```python
boxsize = 4. # 10'

def make_find_chart(starname, filename):
    f = plt.figure(figsize=[5,4.5])
    tbl = pyfits.open(filename)[1].data
    idx = tbl['CATALOG_ID'] == starname
    print '#Target'
    print '%i %10.6f %10.6f %5.2f %5.2f'%(tbl['CATALOG_ID'][idx][0],tbl['RA'][idx][0],tbl['DEC'][idx][0], tbl['WAVG_MAG_PSF_G'][idx][0],tbl['WAVG_MAG_PSF_R'][idx][0])
    plt.plot(tbl['RA'][idx][0],tbl['DEC'][idx][0], marker='*', c='r', ms=20, mec='None',zorder=10)
    box_mask = (abs(tbl['RA'] - tbl['RA'][idx][0]) < boxsize/60/np.cos(tbl['DEC'][idx][0]*np.pi/180)) & (abs(tbl['DEC'] - tbl['DEC'][idx][0]) < boxsize/60)
    bright = (tbl['WAVG_MAG_PSF_G'] < 17) & box_mask
    print '#Bright stars (g<17)'
    for i in range(len(tbl[bright])):
        print '%i %10.6f %10.6f %5.2f %5.2f'%(tbl['CATALOG_ID'][bright][i],tbl['RA'][bright][i],tbl['DEC'][bright][i], tbl['WAVG_MAG_PSF_G'][bright][i], tbl['WAVG_MAG_PSF_R'][bright][i])
        plt.plot(tbl['RA'][bright][i],tbl['DEC'][bright][i], marker='o', c='b', ms=20, mec='None', zorder=5)
    plt.xlim(tbl['RA'][idx][0]+boxsize/60/np.cos(tbl['DEC'][idx][0]*np.pi/180), tbl['RA'][idx][0]-boxsize/60/np.cos(tbl['DEC'][idx][0]*np.pi/180))
    plt.ylim(tbl['DEC'][idx][0]-boxsize/60, tbl['DEC'][idx][0]+boxsize/60)
    plt.xlabel('RA (degree)')
    plt.ylabel('DEC (degree)')
    ax = plt.gca()
    ax.xaxis.set_major_formatter(FormatStrFormatter('%.2f'))
    plt.tight_layout()
    plt.show()
                                                                                               
```

```python
#tuc III
print '#Tucana III'
print '#         ID          RA        DEC      gmag   rmag'
member = [4801599574360614912,4801599574355752448,4801599574505252352]
filename = '/Users/astrolab/Dropbox/dwarfgalaxy/member_prob/tucana_iii_mcmc_v11.3.fits'
for star in member:
    make_find_chart(star,filename)
```
